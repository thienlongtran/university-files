```python
import graphviz
from graphviz import Digraph

#Child PIDS; Parent PIDS
processes = """28681 ; 24527
28683 ; 28681
28692 ; 28687
28682 ; 28681
28687 ; 28684
28685 ; 28681
28686 ; 28682
28689 ; 28682
28693 ; 28683
28690 ; 28681
28694 ; 28686
28695 ; 28685
28691 ; 28684
28684 ; 28682
28688 ; 28683
28696 ; 28688
"""

processList = processes.split("\n")
processList = list(filter(None,processList))
del processList[0] #Takes care of unintended system fork
```

```python
fourForksProcessTree = Digraph(comment="Process Tree - 4 Forks")

for i in processList:
    PID = i.split(" ; ")[0]
    PPID = i.split(" ; ")[1]
    
    fourForksProcessTree.node(PID, "PID: " + PID)
    fourForksProcessTree.node(PPID, "PID: " + PPID)
    fourForksProcessTree.edge(i.split(" ; ")[1], PID)

fourForksProcessTree
```
